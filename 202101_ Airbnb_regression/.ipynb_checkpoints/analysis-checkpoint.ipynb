```python
select_columns = ["accommodates","room_type","review_scores_rating","number_of_reviews","bed_type,bathrooms"]
```

```python
# 引数が"t"ならば1を、"f"ならば0を返す関数change_tf()を作成。
def change_tf(x):
    if x == "t":
        return 1
    elif x == "f":
        return 0

# cleaning_feeとhost_identity_verified,instant_bookableのt,fを1,0に変換してください。
train_data['cleaning_fee'] = train_data["cleaning_fee"].apply(lambda x:change_tf(x))
train_data['host_identity_verified'] = train_data["host_identity_verified"].apply(lambda x:change_tf(x))
train_data['instant_bookable'] = train_data["instant_bookable"].apply(lambda x:change_tf(x))


# 正しく変換されたか確認。
print(train_data[['cleaning_fee','host_identity_verified','instant_bookable']])
```

```python
# cleaning_feeとyの箱ひげ図を描画
sns.boxplot(data=train_data, x="cleaning_fee", y="y")

# 表示範囲を制限
plt.ylim(0, 600)
plt.show()

# host_identity_verifiedとyの箱ひげ図を描画
sns.boxplot(data=train_data, x="host_identity_verified", y="y")

# 表示範囲を制限
plt.ylim(0, 600)
plt.show()

# instant_bookableとyの箱ひげ図を描画
sns.boxplot(data=train_data, x="instant_bookable", y="y")

# 表示範囲を制限
plt.ylim(0, 600)
plt.show()
```

```python
select_columns = ["accommodates","room_type","review_scores_rating","number_of_reviews","bed_type","bathrooms", "cleaning_fee"]

dummy_data = pd.get_dummies(train_data[select_columns], columns=["room_type", "bed_type"], drop_first=True)

# ダミー変数化されていることを確認。
dummy_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
from sklearn.linear_model import LinearRegression

X_train,X_test,y_train,y_test = train_test_split(dummy_data, train_data["y"], random_state = 0)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 偏回帰係数の表示
pd.DataFrame(lr.coef_, index=X_train.columns)
```

```python
# 切片の表示
lr.intercept_
```

```python
# X_testに対する予測値を算出
y_pred_test = lr.predict(X_test)
```

```python
rmse_test = np.sqrt(MSE(y_test, y_pred_test))
```

```python
# RMSEの表示
print(rmse_test)
```

```python
#精度を上げるためにモデルの改善
# room_typeがEntire home/aptのデータのbathroomsとyの箱ひげ図を書く。
sns.boxplot(data=train_data[train_data["room_type"] == "Entire home/apt"], x="bathrooms", y="y")
plt.show()

# room_typeがShared roomのデータのbathroomsとyの箱ひげ図を書く。
sns.boxplot(data=train_data[train_data["room_type"] == "Shared room"], x="bathrooms", y="y")
plt.show()
```

```python
train_data["room_type"].value_counts()
```

```python
# room_type別にデータを分割
data_entire = train_data[train_data["room_type"] == "Entire home/apt"]
data_private = train_data[train_data["room_type"] == "Private room"]
data_share = train_data[train_data["room_type"] == "Shared room"]

# ダミー変数化
dummy_data_entire = pd.get_dummies(data_entire[select_columns], drop_first=True)
dummy_data_private = pd.get_dummies(data_private[select_columns], drop_first=True)
dummy_data_share = pd.get_dummies(data_share[select_columns], drop_first=True)
```

```python
X_train_e,X_test_e,y_train_e,y_test_e = train_test_split(dummy_data_entire, data_entire['y'], random_state = 0)
X_train_p,X_test_p,y_train_p,y_test_p = train_test_split(dummy_data_private, data_private['y'], random_state = 0)
X_train_s,X_test_s,y_train_s,y_test_s = train_test_split(dummy_data_share, data_share['y'], random_state = 0)

model_e = LinearRegression()
model_e.fit(X_train_e, y_train_e)
pred_e_train = model_e.predict(X_train_e)
pred_e_test = model_e.predict(X_test_e)

model_p = LinearRegression()
model_p.fit(X_train_p, y_train_p)
pred_p_train = model_p.predict(X_train_p)
pred_p_test = model_p.predict(X_test_p)


model_s = LinearRegression()
model_s.fit(X_train_s, y_train_s)
pred_s_train = model_s.predict(X_train_s)
pred_s_test = model_s.predict(X_test_s)

# pred_e_train, pred_p_train, pred_s_trainを1つの配列にして、pred_trainに代入してください。
pred_train = np.concatenate([pred_e_train, pred_p_train, pred_s_train])
# 実際の値であるy_train_e,y_train_p,y_train_sを1つの配列にして、answer_trainに代入してください。
answer_train = np.concatenate([y_train_e,y_train_p,y_train_s])
# pred_train,answer_trainから全訓練データの予測値のRMSEを求め、rmse_trainに代入してください。
rmse_train = np.sqrt(MSE(pred_train, answer_train))

# 同様に全テストデータの予測値のRMSEを求め、rmse_testに代入してください。
pred_test = np.concatenate([pred_e_test, pred_p_test, pred_s_test])
answer_test = np.concatenate([y_test_e,y_test_p,y_test_s])
rmse_test = np.sqrt(MSE(pred_test, answer_test))

print(rmse_train)
print(rmse_test)
```
