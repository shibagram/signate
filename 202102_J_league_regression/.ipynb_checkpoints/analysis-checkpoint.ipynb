2. データの前処理

```python
#matchの前処理

#第何節なのかをmatch_numという新たな特徴量に格納
def get_match_num(x):
    return x[x.find("第")+1 : x.find("節")]

data['match_num'] = data["match"].apply(get_match_num)
print(data['match_num'].unique())
```

```python
#gamedayの前処理
def get_month(x):
    return int(x[0]+x[1])

data["month"] = data["gameday"].apply(get_month)
data["month"].unique()
```

```python
#gamedayの前処理
def get_week(x):
    return x[6]

data["week"] = data["gameday"].apply(get_week)
data["week"].unique()
```

```python
#tvの前処理
def get_tv_num(x):
    return len(x.split("／"))

data["tv_num"] = data["tv"].apply(get_tv_num)
data["tv_num"].unique()
```

```python
#humidityの前処理
def get_humidity(x):
    return int(x[:-1])/100

data["humidity"] = data["humidity"].apply(get_humidity)
data['humidity'].unique()
```
